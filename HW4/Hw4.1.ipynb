```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
time = np.linspace(0, 11, num=12)
time
```

```python
plt.scatter(x=time, y=speed)
plt.grid()
plt.xlim([0, 11])
plt.ylim([0, 130])
```

```python
f = scipy.interpolate.interp1d(time, speed, kind='cubic')
more_date = np.linspace(0, 11, 1000)
plt.plot(more_date, f(more_date))
```

```python
result = scipy.integrate.quad(f, 0, 11)
print(result)
```

```python
f_quad = scipy.interpolate.interp1d(time, speed, kind='quadratic')
more_date = np.linspace(0, 11, 1000)
plt.plot(more_date, f_quad(more_date))
```

```python
result = scipy.integrate.quad(f_quad, 0, 11)
print(result)
```
