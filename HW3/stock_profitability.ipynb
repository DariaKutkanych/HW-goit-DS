```python
import matplotlib.pyplot as plt
```

```python
import random

full_x = [0]
full_y = [0]

def add_dot(x, y, num):
    if num == 0:
        full_x.append(x/2)
        full_y.append(y/2)
    elif num == 1:
        full_x.append((x + 1) / 2)
        full_y.append(y/2)
    else:
        full_x.append((x + 0.5) / 2)
        full_y.append((y + 3**0.5) / 2)
        
for b in range(5000):
    num = random.randint(0, 2)
    add_dot(full_x[-1], full_y[-1], num)

plt.scatter(full_x, full_y)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
a = [25, -10, 10, 5, 35, 13]
b = [0, 15, -5, 5, 20, 25]
c = [10, 25, -15, -5, -5, 15]

fig, axs = plt.subplots(ncols=3, figsize=(10, 3))
df = pd.DataFrame([a, b, c], index=["A", "B", "C"]).transpose()
sns.regplot(x="A", y="B", data=df, ax=axs[0])
sns.regplot(x="A", y="C", data=df, ax=axs[1])
sns.regplot(x="B", y="C", data=df, ax=axs[2])
fig.tight_layout()
plt.show()
```
