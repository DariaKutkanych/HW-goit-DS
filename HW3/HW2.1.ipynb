```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"
```

```python
birth_rate = pd.read_html(url)[10]
```

```python
birth_rate.head(10)
```

```python
birth_rate.shape
```

```python
birth_rate_Nan = birth_rate.applymap(lambda x: np.nan if x == "—" else x)
```

```python
birth_rate_Nan.dtypes
```

```python
birth_rate_Nan[["2014", "2019"]] = birth_rate_Nan[["2014", "2019"]].apply(pd.to_numeric)
```

```python
birth_rate_Nan.dtypes
```

```python
birth_rate_Nan.isnull().sum()
```

```python
birth_rate_no_total = birth_rate_Nan.iloc[:-1]
```

```python
birth_rate_no_total.tail(5)
```

```python
birth_rate_na_mean = birth_rate_Nan.apply(lambda x: x.fillna(x.mean() if x.name not in ["Регіон"] else x), axis=0)
```

```python
birth_rate_na_mean
```

```python
%matplotlib inline

data_x = birth_rate_na_mean[["1950", "1960", "1970", "1990", "2000", "2012", "2014", "2019"]][birth_rate_na_mean["Регіон"]=="Україна"]
plt.plot(data_x.iloc[0])
plt.legend(["Ukraine"], loc=1)
plt.xlabel("Years")
plt.ylabel("Avg birth rate")
plt.title("Birth rate in Ukraine, by year")
plt.show()

new = birth_rate_na_mean.transpose()
new.columns = new.iloc[0]
new = new.iloc[1: , :-1]
new.plot(kind='bar', stacked=True,
        title='Total birth rate by year and region')
plt.legend(prop={'size': 4})

plt.show()
```
